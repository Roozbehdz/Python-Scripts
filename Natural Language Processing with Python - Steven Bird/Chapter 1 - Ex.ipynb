Compare the lexical diversity scores for humor and romance fiction in Table
1-1. Which genre is more lexically diverse?

```python
from nltk.corpus import brown
brown.categories()
```

```python
def lexical_diversity(text):
    return len(text)/len(set(text))
```

```python
for genre in list(brown.categories()):
    print(genre)
    print(lexical_diversity(brown.words(categories =genre)))
```
